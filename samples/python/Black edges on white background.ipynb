```python
import cv2
import matplotlib.pyplot as plt
```

```python
path = "misc/clk.jpg"

img = cv2.imread(path, 0)

L1 = cv2.Canny(img, 10, 50, L2gradient=False)

ret, L2 = cv2.threshold(L1, 128, 255, cv2.THRESH_BINARY_INV)


titles = ['Original Image', 'White edges', 'Black edges']

outputs = [img, L1, L2]

plt.figure(figsize=(15, 15))
for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(outputs[i], cmap='gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
